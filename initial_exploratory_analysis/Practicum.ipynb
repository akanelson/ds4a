## Datasets details

### Itinerary Data  
2 Itinerary files:

**Itinerary_dist1_ano.csv**
- Delivered itineraries of agency 1
- Size: 5.26 MB
- Records: 20,992

**Itinerary_dist2_ano.csv**
- Size: 3.19 MB
- Records: 12,425  
- Delivered itineraries of agency 2  

**Columns (in both files): 24**
- itinerary_id (string(32)): itinerary identifier
- driver_id (string(32)): driver identifier
- created (Datetime fmt('%y-%m-%d %H:%M'))
- accepted (Datetime fmt('%y-%m-%d %H:%M'))
- dropped (Datetime fmt('%y-%m-%d %H:%M'))
- started (Datetime fmt('%y-%m-%d %H:%M'))
- finished (Datetime fmt('%y-%m-%d %H:%M'))
- status (string): 9 categories ('finished', 'dropped', 'waiting_pick_up', 'cancelled', 'cancelledWithCharge', 'started', 'allocating', 'requires_verification', 'awaiting_completion')
- total_distance (int64)
- transport_type (string): 2 categories ('moto', 'carro')
- product (string): 1 category ('Pro')
- product_version (string): 2 categories ('Start', 'Prime')
- distribution_center1 (int): 2 category (1, 2)
- packages (int): range 1 to 50
- delivered_packages (int): range 1 to 50
- checked_in_at (Datetime fmt('%y-%m-%d %H:%M'))
- pickup_checkout_at (Datetime fmt('%y-%m-%d %H:%M'))
- pickup_lat (float): latitude range
- pickup_lng (float): longitude range
- real_completion_time (float): looks like in seconds
- pickup_distance (float): looks like in meters
- pickup_time (float): looks like in minutes
- check_in_time (float): looks like in seconds
- waypoints (int): range from 2 to 49 (not null)

We have consolidate itineraries datasets into one postgres table 'itineraries'

## Test a very simple model to predict dropped using drivers and itineraries

````python
df['DOW'] = df.index.dayofweek
````

````python
condition = df.itineraries > 1
df_X, df_y = df[condition][['drivers', 'itineraries', 'DOW']], df[condition][['dropped']]
````

````python
df_X = pd.concat([df_X, pd.get_dummies(df_X['DOW'])], axis=1)
````

````python
df_X = df_X.rename(columns={i : dow[i] for i in range(7)})
del df_X['DOW']
df_X.head()
````

````python
X, y = df_X.to_numpy(), df_y.to_numpy()
dates = df[condition].index.tolist()
````

````python
len(X), len(y)
````

````python
X_train, X_test = X[:100], X[100:]
y_train, y_test = y[:100], y[100:]
dates_train = dates[:100]
dates_test = dates[100:]
````

````python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
````

````python
regr = linear_model.LinearRegression()
````

````python
# Train the model using the training sets
regr.fit(X_train, y_train)
````

````python
# The coefficients
print('Coefficients: \n', regr.coef_)
````

````python
y_pred = regr.predict(X_test)
````

````python
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
````

````python
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
````

````python
df_pred = pd.DataFrame({'Test': y_test.flatten(), 'Pred': y_pred.flatten()}, index=dates_test)
````

````python
df_pred.plot(figsize=(15,6));
plt.xticks(rotation=90);
````

**Our simple prediction does not look so bad at all!**

There is a big peak not predicted, but this is just the begining! It is pending look for holidays relations, traffic, etc.
